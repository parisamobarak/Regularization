```python
import pandas as pd    #رگرسیون خطی با تنظیم
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

df = pd.read_excel('dataset.xlsx', header=None)
df = df.values
print("dataset:\n", df)

x, y = df[:, 0:-1], df[:, -1]


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

def h(x, theta):
    return x.dot(theta)

def mse(y_pred, y_true):  #تغییر تابع خطا
    regularization_l2 = lambda_ * np.sum(theta[1:]**2)  
    return 0.5 * ((y_pred - y_true) ** 2).mean() + regularization_l2

mu = x_train.mean(axis=0)
sigma = x_train.std(axis=0)
xn_train = (x_train - mu) / sigma

alpha = 0.01
lambda_ = 0.1 # تنظیم مقدار lambda_

theta = np.random.randn(x.shape[1] + 1)
costs_train = []
xn_train = np.insert(xn_train, 0, 1, axis=1)

def update_step():
    global theta, costs_train
    y_pred = h(xn_train, theta)
    costs_train.append(mse(y_pred, y_train))
    dtheta = np.zeros([xn_train.shape[0], xn_train.shape[1]])
    dtheta[:, 0] = (y_pred - y_train)
    
    for i in range(1, xn_train.shape[1]):
        dtheta[:, i] = xn_train[:, i] * dtheta[:, 0]
        dtheta[:, i] += lambda_ * theta[i]  # تغییر در دلتاتتا های غیرثابت 
        
    for i in range(0, xn_train.shape[1]):
        theta[i] -= alpha * dtheta[:, i].mean()

costs_test_with_regularization = []  # خطاهای تست با تنظیم L2
for i in range(1000): 
    update_step()
    x_test_n = (x_test - mu) / sigma
    x_test_n = np.insert(x_test_n, 0, 1, axis=1)
    y_pred_test = h(x_test_n, theta)
    costs_test_with_regularization.append((mse(y_pred_test, y_test))) #اضافه کردن خطاهای با تنظیم به لیست
    
plt.plot(costs_test_with_regularization, label='With L2 Regularization')  #رسم نمودار خطاهای با تنظیم
theta_with_regularization=theta # ذخیره پارامترهای با تنظیم   



#انجام الگوریتم بدون تنظیمl2

def mse1(y_pred , y_true):
    return 0.5*((y_pred - y_true)**2).mean()
mu = x_train.mean(axis=0)
sigma = x_train.std(axis=0)
xn_train = (x_train - mu) / sigma
theta = np.random.randn( x.shape[1] +1 )
costs_train=[]
xn_train = np.insert(xn_train, 0, 1, axis=1)
def update_step1():
    global theta , costs_train
    y_pred= h(xn_train ,theta)
    costs_train.append(mse1(y_pred , y_train))
    dtheta = np.zeros([xn_train.shape[0],xn_train.shape[1]])
    dtheta[:,0]=(y_pred - y_train)
    for i in range(1,xn_train.shape[1]):
        dtheta[:,i]= xn_train[: ,i]*dtheta[:,0]      
    for i in range(0,xn_train.shape[1]):
        theta[i]-=alpha*dtheta[: ,i].mean()

costs_test_without_regularization = []  # خطاهای تست بدون تنظیم

for i in range(1000): 
    update_step1()
    x_test_n = (x_test - mu) / sigma
    x_test_n = np.insert(x_test_n, 0, 1, axis=1)
    y_pred_test = h(x_test_n, theta)
    costs_test_without_regularization.append((mse1(y_pred_test, y_test))) #اضافه کردن خطاهای بدون تنظیم به لیست

theta_without_regularization=theta  #ذخیره پارامترهای بدون تنظیم  

plt.plot(costs_test_without_regularization, label='Without  L2 Regularization') #رسم نمودار خطای بدون تنظیم
plt.xlabel('iteration')
plt.ylabel('costs')
plt.legend()
plt.show()



# مقایسه مقادیر پارامترها
print("Parameters with L2 Regularization:", theta_with_regularization)
print("Parameters without L2 Regularization:", theta_without_regularization)
```

```python
import pandas as pd   #رگرسیون خطی با رابطه نرمال با تنظیم 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

df = pd.read_excel('dataset.xlsx', header=None)
df = df.values
print("dataset:\n", df)

x, y = df[:, 0:-1], df[:, -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

def h(x, theta):
    return x.dot(theta)

def mse(y_pred, y_true, theta, lambda_):
    regularization_l2 = lambda_ * np.sum(theta[1:]**2)  # جمع مجموعه پارامترهای تتا با تنظیم L2
    return 0.5 * ((y_pred - y_true) ** 2).mean() + regularization_l2


lambda_ = 0.1  


mu = x_train.mean(axis=0)
sigma = x_train.std(axis=0)
xn_train = (x_train - mu) / sigma


xn_train = np.insert(xn_train, 0, 1, axis=1)


I = np.eye(xn_train.shape[1])
I[0, 0] = 0 
theta_with_regularization = np.linalg.pinv(xn_train.T.dot(xn_train) + lambda_ * I).dot(xn_train.T).dot(y_train)


x_test_n = (x_test - mu) / sigma
x_test_n = np.insert(x_test_n, 0, 1, axis=1)

y_pred_test_with_regularization = h(x_test_n, theta_with_regularization)
costs_test_with_regularization = mse(y_pred_test_with_regularization, y_test, theta_with_regularization, lambda_)

# انجام الگوریتم بدون تنظیم

theta_without_regularization = np.linalg.pinv(xn_train.T.dot(xn_train)).dot(xn_train.T).dot(y_train)

y_pred_test_without_regularization = h(x_test_n, theta_without_regularization)
costs_test_without_regularization = mse(y_pred_test_without_regularization, y_test, theta_without_regularization, 0)  # با L2 = 0

# چاپ خروجی‌ها
print("Cost with L2 Regularization:", costs_test_with_regularization)
print("Cost without L2 Regularization:", costs_test_without_regularization)

# مقایسه پارامترها
print("Parameters with L2 Regularization:", theta_with_regularization)
print("Parameters without L2 Regularization:", theta_without_regularization)
```
